```python
#merge columns using outer merge to retain alll the date 
frames =[df_appstore1, df_appstore2, df_appstore3, df_appstore4]
df_merged = pd.concat(frames)
df_merged.head()
```

```python
dup_rows = df_merged.duplicated()
print(dup_rows)
```

```python
missing_values = df_merged.isna().sum()
print(missing_values)
```

```python
#drop columns that we are not going to use 
df_merged.drop(columns=["Version"], inplace = True)
```

```python
df_merged.columns
```

```python
df_merged.rename(columns ={'Prefrence':'Preference'}, inplace = True)
df_merged.rename(columns ={'Avy Rating (Current)':'Avg Rating (Current)'}, inplace = True)
df_merged.columns
```

```python
df_merged.dtypes
```

```python
import re

# Function to remove special characters from a string
def rem_spec_char(text):
    return re.sub(r'[^a-zA-Z0-9\s]+', '', text)


df_merged['App Name'] = df_merged['App Name'].apply(rem_spec_char)
```

```python
df_merged.head()
```

```python
df_merged.tail()
```

```python
def bytes_to_megabytes(bytes):
    return bytes / (1024 * 1024)

df_merged['Size (MB)'] = round(df_merged['Size (bytes)'].apply(bytes_to_megabytes))

df_merged.head()
```

```python
#Summary Statistics for the price
med_price = df_merged['Price'].median()
mean_price = df_merged['Price'].mean()
print("Median Price:", med_price, 'USD')
print("Average Price:", round(mean_price ,2),'USD')
```

```python
#Summary Statistics for Supp Devices 
avg_supp_dev = df_merged['Supp Devices'].mean()
print("Average supported devices: ", round(avg_supp_dev))
```

```python
#Summary Statistics for supported languages 
avg_supp_lang = df_merged['Supp Lang'].mean()
print("Average number of supported languages:", round(avg_supp_lang))
```

```python
import seaborn as sns
df_grp_data = df_merged.groupby('Preference')[['Supp Devices', 'Screenshot', 'Supp Lang']].mean()
sns.heatmap(df_grp_data.corr())
```

### There is a positive correlation between Supported Devices and Screenshot as well as Screenshot and the number of supported devices this suggests that there is an association among the attributes as one increases so does the other.
### The data is grouped by 'Preference', the correlations may vary across different preference categories. This will be investigated further in the next figure

```python
grouped_data = df_merged.groupby('Preference')

# Loop through each group
for group_name, group_data in grouped_data:
    # Calculate correlation matrix for the group
    corr_matrix = group_data[['Supp Devices', 'Screenshot', 'Supp Lang']].corr()
    
    # Plot heatmap for the correlation matrix
    plt.figure(figsize=(8, 6))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title(f'Correlation Heatmap for {group_name}')
    plt.show()
```

### Upon investigating this further to a granular level, it can be seen that the data shows no clear pattern or trend, with significant variability observed across various app categories
### In some cases there is a moderate relationship between supported languages and screenshot.
